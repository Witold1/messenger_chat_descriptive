## Summary: <a class="anchor" id="sum"></a>

> Для анализа даны распределенные во времени 2852 сообщения двух человек, из которых 2811 авторские сообщения или скопированные сообщения и 41 replay. Из них 68 фотографий, 23 уникальных стикера, 3 голосовых сообщения, 1 видеофайл, 1 .pdf файл. Так как данные по природе своей персональные и защищены законом о тайне переписки (138 УК РФ) и здравым смыслом, сырой набор данных, как и код обработки не опубликован.  
<sup>\* Погрешность в оценке количества сообщений не превышает 50 сообщений.</sup>  
<sup>\*\* Погрешность сборf данных не оценена, так как некоторые сообщения могли удаляться и логи удаления источником не ведутся.</sup>

> Временной период ряда: 135 дней. Количество дней без общения (пропуски) 17 дней или 12.59% всего периода. Зрительным анализом выявлен гасящий тренд и возможная дневная и недельная сезонность.  
<sup>\* Погрешность не оценена. Значение пропусков в не строгом интервале от 17 дней до 25 дней (19%). </sup>

> Лингвистический описательный анализ на большей части выборки паттернов не выдал, но есть пересекающиеся у двух человек слова (например, представление). 

> Резюмируя дальнейший анализ, с начала периода к концу периода люди стали общаться **меньше**.

**Отношение количества сообщений по людям, динамика которого показана выше**: <a class="anchor" id="ratio"></a> 

* В 58% дней (примерно 78 дней) отношение больше 1 (Я писал больше).
* В 16% дней (примерно 22 дня) отношение равно 1 (равное количество сообщений).
* В 7% дней (примерно 10 дней) отношение меньше 1 (Ты писала больше)

Медианное значение равно отношения при не учете пропусков равно 1.13, при замене пропусков нулями 1.08. То есть, Я писал больше, но не намного.
